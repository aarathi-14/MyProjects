```
x, y = make_blobs(n_samples=2223, centers=5 , cluster_std=0.60, random_state=0)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```
#Considering K value is 5
kmeans = KMeans(n_clusters=5)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=30)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100)
plt.show()
```

```
kmeans = KMeans(n_clusters=5).fit(x)
KMeans(n_clusters=5).fit(x)
normalized_vectors = preprocessing.normalize(x)
normalized_kmeans = KMeans(n_clusters=5).fit(normalized_vectors)
print("Values for k=5")
print('Kmeans: {}'.format(silhouette_score(x, kmeans.labels_, metric='euclidean')))
```

```
#Considering K value is 4
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=30)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100)
plt.show()
```

```
kmeans = KMeans(n_clusters=4).fit(x)
KMeans(n_clusters=4).fit(x)
normalized_vectors = preprocessing.normalize(x)
normalized_kmeans = KMeans(n_clusters=4).fit(normalized_vectors)
print("Values for k=4")
print('Kmeans: {}'.format(silhouette_score(x, kmeans.labels_, metric='euclidean')))
```

```
#Finding optimal K value
optimalK =[] 
for i in range(1, 11): 
    KM = KMeans(n_clusters = i) 
    KM.fit(X) 
    optimalK.append(KM.inertia_)       
plt.plot(range(1, 11), optimalK, color ='blue', linewidth ='1') 
plt.xlabel("K value") 
plt.ylabel("Error") 
plt.show()
```
